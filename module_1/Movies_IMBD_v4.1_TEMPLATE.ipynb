# 24. Какая студия дает самые длинные названия своим фильмам по количеству символов?

```python
answers['24'] = 'Four By Two Productions'
```

```python
data24 = data.copy()

data24['title_length'] = data24['original_title'].str.len() #создаем столбец с количеством символов в названиях

data24['production_companies'] = data24['production_companies'].str.split('|') #разбиваем на отдельные компании
data24 = data24.explode('production_companies') #разносим по отдельным строкам
#группируем по компаниям, считаем среднее количество символов и смотрим наибольшее
data24.groupby(['production_companies']).mean()['title_length'].sort_values(ascending=False)
```
